```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
academic_performance_df=pd.read_csv('academic_performance_dataset_V2.csv')
lamborghini_sales_df=pd.read_csv('lamborghini_sales_2020_2025.csv')
```

```python
print("Academic_performance:")
print(academic_performance_df.head())
print("\nLamborghini_sales:")
print(lamborghini_sales_df.head())
```

```python
print("Academic_performance:")
print(academic_performance_df.tail())
print("\nLamborghini_sales:")
print(lamborghini_sales_df.tail())
```

```python
print("Academic_performance:")
print(academic_performance_df.describe())
print("\nLamborghini_sales:")
print(lamborghini_sales_df.describe())
```

```python
print("Academic_performance:")
print(academic_performance_df.dropna())
print("\nLamborghini_sales:")
print(lamborghini_sales_df.dropna())
```

```python
print("Academic_performance:")
print(academic_performance_df.isnull().sum())
print("\nLamborghini_sales:")
print(lamborghini_sales_df.isnull().sum())
```

```python
print("Academic_performance:")
academic_performance_df.plot()
plt.show()
print("\nLamborghini_sales:")
lamborghini_sales_df.plot()
plt.show()
```
